```python
from torchvision import transforms
import torch
import numpy as np
from PIL import Image

# ✅ Function to preprocess image correctly
def preprocess_image(image_path):
    image = Image.open(image_path).convert("RGB")

    # ✅ Auto-crop to remove excess whitespace (focus on handwriting)
    image = np.array(image)
    gray = np.mean(image, axis=2)  # Convert to grayscale
    mask = gray < 200  # Detect text (darker than white background)
    coords = np.argwhere(mask)  # Get text coordinates
    if coords.shape[0] > 0:
        x_min, y_min = coords.min(axis=0)
        x_max, y_max = coords.max(axis=0)
        image = Image.fromarray(image[y_min:y_max, x_min:x_max])  # Crop text area

    # ✅ Resize while keeping aspect ratio
    transform = transforms.Compose([
        transforms.Resize((384, 384)),  # Resize to match TrOCR input
        transforms.ToTensor(),
        transforms.Normalize((0.5,), (0.5,))
    ])
    image = transform(image)

    return image.unsqueeze(0)  # Add batch dimension

print("✅ Image Preprocessing Fixed!")
```

```python
import cv2
import pytesseract
from pdf2image import convert_from_path
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
import os

# Define function for OCR from image
def extract_text_from_image(image_path):
    """Extract text from an image using OCR."""
    image = cv2.imread(image_path)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # Convert to grayscale
    text = pytesseract.image_to_string(gray)  # OCR Extraction
    return text

# Function to process PDF pages into images and extract text
def extract_text_from_pdf(pdf_path):
    """Extract text from a PDF by converting pages to images."""
    poppler_path = "/usr/bin" # Poppler executable path
    if not os.path.exists(os.path.join(poppler_path, "pdfinfo")):
         raise Exception("Poppler is not correctly installed. Please install poppler-utils")
    images = convert_from_path(pdf_path, poppler_path=poppler_path)
    extracted_text = ""

    for i, img in enumerate(images):
        img.save(f"page_{i}.jpg", "JPEG")  # Save each page as an image
        text = extract_text_from_image(f"page_{i}.jpg")
        extracted_text += text + "\n\n"

    return extracted_text

# Upload Exam Sheet (Image/PDF) in Colab
from google.colab import files

uploaded = files.upload()  # Allows user to upload an image or PDF

# Extract text based on file type
for filename in uploaded.keys():
    if filename.endswith(".pdf"):
        extracted_text = extract_text_from_pdf(filename)
    else:
        extracted_text = extract_text_from_image(filename)

    print("\nExtracted Text:\n", extracted_text)
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

def evaluate_answer(question, student_answer):
    """Evaluate student answer without comparing to a correct answer"""
    question_embedding = model.encode(question, convert_to_tensor=True)
    student_embedding = model.encode(student_answer, convert_to_tensor=True)

    similarity_score = util.pytorch_cos_sim(question_embedding, student_embedding).item()

    score = round(similarity_score * 10, 2)

    if score > 8:
        feedback = "Excellent response! Well articulated."
    elif score > 5:
        feedback = "Good attempt! Some details could be improved."
    else:
        feedback = "Needs improvement. Try elaborating more on the concept."

    return {"Score": f"{score}/10", "Feedback": feedback}

question = "What is machine learning?"
student_answer = "Machine learning is about AI learning from experience."

result = evaluate_answer(question, student_answer)

print("\nEvaluation Result:\n", result)
```
